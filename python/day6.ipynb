[문제 2] 여러 숫자를 입력받아 그 합계를 반환하는 함수 sum_numbers를 작성하세요. 이 때, 가변 매개변수를 사용하세요.

```python
def sum_numbers(n, *values) -> int:
    """주어진 숫자들의 합을 반환합니다."""
    total = n
    for value in values:
        total += value
    return total
input_values = input("숫자들을 공백으로 구분하여 입력하세요: ")
numbers = list(map(int, input_values.split()))
result = sum_numbers(*numbers)      
print(f"합계는: {result}")
```

[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n: int) -> list[int]:
    """0부터 n까지의 홀수 리스트를 반환합니다."""
    return [i for i in range(n) if i % 2 != 0]
```

[문제 8] 사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
input_values = input("세 숫자를 공백으로 구분하여 입력하세요: ")
numbers = tuple(map(int, input_values.split()))
def min_max(*args):
    """주어진 숫자들의 최소값과 최대값을 반환합니다."""
    return min(args), max(args)
minimum, maximum = min_max(*numbers)
print(f"최소값: {minimum}, 최대값: {maximum}")
```

[문제 9] 주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(n: int) -> bool:
    """주어진 숫자가 짝수인지 여부를 반환합니다."""
    if n % 2 == 0:
        return True
    else:
        return False
    
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def fac_recursive(n: int) -> int:
    """재귀적으로 n의 팩토리얼을 계산합니다."""
    if n < 0:
        return ValueError("음수의 팩토리얼은 정의되지 않습니다.")
    if n == 0 or n == 1:
        return 1
    return n * fac_recursive(n - 1)     
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(strings: list[str]) -> None:
    """주어진 문자열들의 길이를 출력합니다."""
    for s in strings:
        print(f"'{s}'의 길이: {len(s)}")    
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(s1: str, s2: str) -> str:
    """두 문자열 중 더 긴 문자열을 반환합니다."""
    if len(s1) >= len(s2):
        return s1
    else:
        return s2
```
